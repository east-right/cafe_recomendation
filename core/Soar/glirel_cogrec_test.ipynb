```python
# pip install glirel spacy
# python -m spacy download ko_core_news_sm

import spacy
from glirel import GLiREL

# 1. 모델 및 한국어 토크나이저 로드
model = GLiREL.from_pretrained("jackboyla/glirel-large-v0")
nlp = spacy.load("ko_core_news_sm")

text = "카카오의 창업자는 김범수이다."
tokens = [token.text for token in nlp(text)]

# 2. 개체명(NER) 정보 입력 (보통 GLiNER 등을 통해 미리 추출해 둡니다)
# 형식: [시작 인덱스, 끝 인덱스, 개체 타입, 텍스트]
ner_results = [
    [0, 1, "ORG", "카카오"], 
    [3, 4, "PERSON", "김범수"]
]

# 3. 찾고자 하는 한국어 관계 레이블 정의
labels = ["창업자", "소속", "경쟁사", "관계없음"]

# 4. 관계 예측
relations = model.predict_relations(
    tokens, 
    labels, 
    threshold=0.0, 
    ner=ner_results, 
    top_k=1
)

# 결과 출력
for rel in relations:
    print(f"{rel['head_text']} -> {rel['label']} -> {rel['tail_text']} (점수: {rel['score']:.4f})")
```

```python
from gliner2 import GLiNER2

# Load the model
extractor = GLiNER2.from_pretrained("fastino/gliner2-multi-v1")

# Extract entities
text = "카공하기 좋은 카페"
ner = ["목적", "평가"]
result = extractor.extract_entities(text, ner)

print(result)
# Output: {'entities': {'company': ['Apple'], 'person': ['Tim Cook'], 'product': ['iPhone 15'], 'location': ['Cupertino']}}
```
